## Variable ranges:

**Score**- The scores are based on answers to the main life evaluation question asked in the poll. This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale. Ranges from 2.9 to 7.63 in this dataset- although 5 is still considered the middle score

**GDP**- ranges from 0 to 2.1

**Social Support**- ranges from 0 to 1.64

**Healthy Life Expectancy**- ranges from 0 to 1.03

**Freedom to Make Life Choices**- ranges from 0 to 0.72

**Generosity**- ranges from 0 to 0.6

**Perceptions of corruption** means the absence of government corruption, higher scores are better

The following columns: GDP per Capita, Family, Life Expectancy, Freedom, Generosity, Trust Government Corruption **describe the extent to which these factors contribute in evaluating the happiness in each country.**

```python
df['Above or below 5'] = np.where(df['Score']> 5 , 'above', 'below')
df
#although the average happiness score in the dataset is a little bit above 5, I wanted to seperate the dataset into above and below 5, which is the middle score a country can get
#I did this to be able to use the groupby function, and simply to see how many countries are considered above and below average happiness
```

```python
df2['Above or below 5'] = np.where(df2['Score']> 5 , 'above', 'below')
df2
```

```python
dftable= df.groupby('Above or below 5').agg({
    'Overall rank': 'count', 
    'GDP per capita': 'mean',
    'Social support': 'mean',
    'Healthy life expectancy':'mean',
    'Freedom to make life choices': 'mean',
    'Generosity': 'mean',
    'Perceptions of corruption': 'mean'
})
dftable
#here I was able to find the differences between the above 5 and below 5 group, but even more helpful is the table below so we can compare how things have changed from 2018 to 2019
```

```python
df2table= df2.groupby('Above or below 5').agg({
    'Overall rank': 'count', 
    'GDP per capita': 'mean',
    'Social support': 'mean',
    'Healthy life expectancy':'mean',
    'Freedom to make life choices': 'mean',
    'Generosity': 'mean',
    'Perceptions of corruption': 'mean'
})
df2table
#here it looks like the count of above and below 5 is the same from 2018, which surprised me as that means it is either a coincidence or there is something I am not understanding about the score variable
#it looks like the wealth gap has widened (unsurprisingly) GDP has increased for the "above average" countries and decreased for below average countries
#This gap widening is the same for social support
#healthy life expectancy has increased for both above and below 5 (yay)
#freedom to make life choices has decreased for both
#generosity also increased across the board (yay again)
#the gap between perceptions/absence of corruption is narrowing (high scores are better), it got worse for above 5 and better for below 5
```
